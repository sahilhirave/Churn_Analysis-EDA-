#1. Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

#2. Loading Data


```python
df = pd.read_csv('Customer_Data.csv')
df.head()
```

##3. Inspection of data

```python
df.info()
```

##4. Replacing empty spaces with zero

```python
df['TotalCharges'] = df['TotalCharges'].replace(' ','0')
df['TotalCharges'] = df['TotalCharges'].astype('float')
```

##5. To check whether data has null values

```python
df.isnull()
```

```python
df.isnull().sum()
```

##6. Convt '0' to False and '1' to True

```python
def covt(value):
  if value == 1:
    return "Yes"
  else:
    return "No"

df['SeniorCitizen'] = df['SeniorCitizen'].apply(covt)
```

##7. Plotting some data

```python
ax = sns.countplot(x='Churn', data=df)
ax.bar_label(ax.containers[0])
plt.title('Count of Customers Churn')
plt.show()
```

# PiePlot for percentage plotting

```python
plt.figure(figsize=(4,4))
gb = df.groupby('Churn').agg({'Churn':'count'})
plt.pie(gb['Churn'],labels=gb.index, autopct = "%1.2f%%")
```

#Creating Stacked Plotting for better analysis of data

```python
# 1. Prepare the data: Group, count, and unstack
plot_data = df.groupby(['SeniorCitizen', 'Churn']).size().unstack()

# 2. Calculate percentages (normalize by row)
plot_data_pct = plot_data.div(plot_data.sum(axis=1), axis=0) * 100

# 3. Plotting
ax = plot_data_pct.plot(kind='bar', stacked=True, figsize=(8, 6), color=['#66b3ff','#ff9999'])

# 4. Adding the percentage labels
for p in ax.patches:
    width = p.get_width()
    height = p.get_height()
    x, y = p.get_xy()
    if height > 0:
        ax.text(x + width/2,
                y + height/2,
                f'{height:.1f}%',
                ha='center',
                va='center',
                fontweight='bold',
                color='white')

plt.title('Churn Percentage by Senior Citizen Status')
plt.ylabel('Percentage (%)')
plt.xlabel('Senior Citizen (0 = No, 1 = Yes)')
plt.xticks(rotation=0)
plt.legend(title='Churn', loc='upper right')
plt.show()
```

##Histogram for Tenure/Churn plotting

```python
plt.figure(figsize=(10, 5))

sns.histplot(
    x='tenure',
    data=df,
    bins=30,
    hue='Churn',
    palette={'Yes': 'crimson', 'No': 'steelblue'},
    alpha=0.6
)

plt.title('Basic Histogram: Tenure by Churn Status')
plt.xlabel('Tenure (Months)')
plt.ylabel('Count')
plt.legend(title='Churn')
plt.show()
```

##Plotting based on Contract

```python
plt.figure(figsize=(8, 5))

sns.countplot(
    x='Contract',
    data=df,
    hue='Churn',
    palette={'Yes': 'crimson', 'No': 'steelblue'}
)

plt.title('Churn Count by Contract Type')
plt.xlabel('Contract Type')
plt.ylabel('Count')
plt.legend(title='Churn')
plt.show()
```

```python
cols = [col for col in df.columns
        if df[col].dtype == 'object' and col != 'customerID']

cols += ['SeniorCitizen']

n_cols = 4
n_rows = (len(cols) + n_cols - 1) // n_cols

fig, axes = plt.subplots(n_rows, n_cols, figsize=(22, n_rows * 4))
axes = axes.flatten()

for i, col in enumerate(cols):

    # Grouped counts
    grouped = df.groupby([col, 'Churn']).size().unstack(fill_value=0)

    categories = grouped.index.astype(str)
    x = np.arange(len(categories))
    width = 0.35

    axes[i].bar(x - width/2, grouped['No'], width, label='No')
    axes[i].bar(x + width/2, grouped['Yes'], width, label='Yes')

    axes[i].set_xticks(x)
    axes[i].set_xticklabels(categories, rotation=45)
    axes[i].set_title(f'{col} vs Churn')
    axes[i].legend()

for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```
